# latticeを作成

```python
lattice=[]
for i in range(len(sentence)):
    lattice.append([])
lattice[0].append([-1,0,1])
```

# lattice処理

```python
for num,i in enumerate(sentence):
    #以下パッチ処理
    if(len(lattice[num-1])==0 and num-1>0):
        for m in lattice[num-2]:
            if(num2status[m[1]]+"-"+j in bigram_prob):
                lattice[num-1].append([m[1],status2num["NN"],m[2]*bigram_prob[num2status[m[1]]+"-"+j]*0.00000001])#中身がないときの処理
            else:
                lattice[num-1].append([m[1],status2num["NN"],m[2]*0.01*0.01])
    #パッチ処理
    for j in status:
        if(i+"/"+j in lex_prob):
            compare_list=[]#確率を比較するためのリスト
            for k in lattice[num-1]:
                pre_word=status[k[1]]
                if(pre_word+"-"+j in bigram_prob):
                    compare_list.append([status2num[pre_word],status2num[j],k[2]*float(bigram_prob[pre_word+"-"+j])*float(lex_prob[i+"/"+j])])#比較するために一時保存用リスト
            if(len(compare_list)!=0):
                cost_compare=[]#コスト比較用リスト
                for l in compare_list:
                    cost_compare.append(l[2])
                cost_compare=np.array(cost_compare)#確率が最大の配列番号
                lattice[num].append(compare_list[np.argmax(cost_compare)])#確率が最大のものをlatticeに追加
            #if(len(lattice[num])):
                #print(lattice)
print(lattice)
                
```

# 最後尾のlatticeのコストを比較し最大確率のlatticeを求める

```python
compare_cost=[]#確率を比較するためのリスト
for i in lattice[len(lattice)-1]:
    compare_cost.append(i[2])
last_word_class=np.argmax(np.array(compare_cost))#確率を比較し最後尾のlatticeを決定
wordclass=[sentence[len(sentence)-1]+"/"+num2status[lattice[len(lattice)-1][last_word_class][1]]]
pointer=lattice[len(lattice)-1][last_word_class][0]#文末からのpointer
```

# 後ろから文頭までポインタをたどる

```python
for i in range(len(lattice)-1):
    decide_list_num=len(lattice)-i-2
    for j in lattice[decide_list_num]:
        if(j[1]==pointer):
            wordclass.append(sentence[decide_list_num]+"/"+num2status[pointer])
            pointer=j[0]
print(wordclass)
```
